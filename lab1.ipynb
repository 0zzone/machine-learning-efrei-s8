### Question 8

To avoid training on redundant features, we are going to remove some high correlated pairs of data features (e.g with correlation > 0.8).

```python
df3 = df.copy()

threshold = 0.8

corr_matrix = df3.corr().abs()

np.fill_diagonal(corr_matrix.values, 0)

high_corr_pairs = np.where(corr_matrix > threshold)

columns_to_drop = set()

for i, j in zip(*high_corr_pairs):
    if i < j:
        colname = corr_matrix.columns[i]
        columns_to_drop.add(colname)

df3 = df3.drop(columns=columns_to_drop)

plt.figure(figsize=(20, 16))
correlation = df3.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
```

We are also going to remove outliers from the dataset, so that the data is clean and ready for machine learning.

```python
from sklearn.preprocessing import StandardScaler


df4 = df3.copy()

c_col = df['c']
df4 = df4.drop('c', axis=1)

Q1 = df4.quantile(0.25)
Q3 = df4.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

final_df = df4[~((df4 < lower_bound) | (df4 > upper_bound)).any(axis=1)]

# Normalize & standardize again 
scaler = MinMaxScaler()
final_df = pd.DataFrame(scaler.fit_transform(final_df), columns=final_df.columns)
scaler = StandardScaler()
final_df = pd.DataFrame(scaler.fit_transform(final_df), columns=final_df.columns)
final_df["c"] = c_col
```

## Part 2 - Train-Test Split and Baseline Model

### Question 1

We use a stratified train-test data split. 

```python
from sklearn.model_selection import train_test_split

X = final_df.drop(columns=['c'])
y = final_df['c']

# Assuming 'y' is the target variable from earlier cells
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, 
    stratify=y, 
    random_state=42
)

final_df
```

### Question 2

We train a baseline Decision Tree. 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 

# Train a baseline Decision Tree
dt_baseline = DecisionTreeClassifier(random_state=42)
dt_baseline.fit(X_train, y_train)
```

### Question 3

We report the accuracy on both training and testing datasets.

```python
print(f"DT Accuracy on Training data: {accuracy_score(y_train, dt_baseline.predict(X_train)):.4f}")

print(f"DT Accuracy on Testing data: {accuracy_score(y_test, dt_baseline.predict(X_test)):.4f}")
```

### Question 4

We report precision, recall, F1 and the confusion matrix on the testing set. 

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

y_pred = dt_baseline.predict(X_test)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(cm)
```

### Question 5

We report any issue of overfitting, underfitting or any other issue. 

```python
train_acc = accuracy_score(y_train, dt_baseline.predict(X_train))
test_acc = accuracy_score(y_test, dt_baseline.predict(X_test))

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Testing Accuracy: {test_acc:.4f}")

print("Training accuracy is significantly higher than testing accuracy.")
print("Therefore, the model may be overfitting: high training accuracy but lower test accuracy.")
```

## Part 3 - Resampling with SMOTE

Objective: Understand class imbalance handling using SMOTE

### Task 1

> Based on the class distribution, do we need under-sampling or oversampling? Justify your choice.

```python
print("Class distribution:\n", y.value_counts())
```

Based on the instructions above, there is a class imbalance. Thus, oversampling is preferred to increase the representation of the minority class without discarding data from the majority class.

### Task 2

> Apply SMOTE using the imblearn.over_sampling module

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# Separate target and features
y = final_df['c'].astype(bool)
X = final_df.drop(columns='c')

# Apply SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Convert back to DataFrame and Series
X_res_df = pd.DataFrame(X_res, columns=X.columns)
y_res_series = pd.Series(y_res, name='c').astype(bool)

# Combine into a single DataFrame
final_df = pd.concat([X_res_df, y_res_series], axis=1)
```

### Task 3

> Show class distribution before and after.

```python
print("Initial class distribution:\n", y.value_counts())
print("Resampled class distribution:\n", y_res.value_counts())
```

### Task 4

> Is there any risk with SMOTE? Data leakage?

SMOTE creates samples data to the undersampling category. Therefore, we can face different issues:
- Some data can be out of bounds, because of outliers. 
- Some other data can be "fake" and not real. Thus, the model may not accurately represent our original data. 

### Task 5

> Are there other oversampling methods?

We could perform random oversampling. This would duplicate samples instead of creating new synthetic ones.

### Exercise 2 - Decision Tree with Pre-Pruning

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, f1_score

model = DecisionTreeClassifier(random_state=42)

param_grid = {
    'max_depth': [3, 5, 10, 20, 30, None],
    'min_samples_split': [5, 10, 20, 30]
}

scoring = 'f1_macro'

grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring=scoring,
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best f1_macro score:", grid_search.best_score_)

grid_search
```

We can notice that we got a F1 score of 0.81 which is actuallt greater than the baseline.
By readin the documentation, the default values for the DecisionTree model is None (max_depth) and 2 (min_samples_split).
Here, we found that the best result has been caused with 20 (max_depth) and 5 (min_samples_split).

### Exercise 3 - Post-Pruning

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt


clf = DecisionTreeClassifier(random_state=42, max_depth=20, min_samples_split=5) # We add the optimal parameters from the grid search
clf.fit(X_train, y_train)

path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.figure()
plt.plot(ccp_alphas, impurities, marker='o', drawstyle="steps-post")
plt.xlabel("ccp_alpha")
plt.ylabel("Total Impurity of Leaves")
plt.title("Cost-Complexity Pruning Path")
plt.show()
```

```python
param_grid = {'ccp_alpha': ccp_alphas[::5]}  # Sample every 5th to reduce computation

grid_search = GridSearchCV(
    estimator=clf,
    param_grid=param_grid,
    scoring='f1_macro',
    cv=5
)

grid_search.fit(X_train, y_train)
print("Best ccp_alpha:", grid_search.best_params_['ccp_alpha'])
print("Best f1_macro:", grid_search.best_score_)
```

We also can use GridSearch to find the optimal ccp_alpha!

To conclude, here we got the best result (over 81% of precision).
However, we did not get a significant difference with the previous question because the default ccp_alphas is 0 for the Decision Tree and the best parameter found is almost 0.
